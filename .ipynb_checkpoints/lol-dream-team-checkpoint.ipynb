```python
import pandas as pd
```

```python
df = pd.read_json('league-of-legends/champion_info_2.json')
obj = []
for item in df["data"]:
    obj.append(item)
champs = pd.DataFrame(obj)
champs = champs[champs["id"]>0].set_index("id")
```

```python
games = pd.read_csv('league-of-legends/games.csv')
```

```python
t1Wins = games[games['winner'] == 1][['t1_champ1id','t1_champ2id','t1_champ3id','t1_champ4id','t1_champ5id']]
t1Wins.columns = ['champ1','champ2','champ3','champ4','champ5']
```

```python
t2Wins = games[games['winner'] == 2][['t2_champ1id','t2_champ2id','t2_champ3id','t2_champ4id','t2_champ5id']]
t2Wins.columns = ['champ1','champ2','champ3','champ4','champ5']
```

```python
wins = pd.concat([t1Wins,t2Wins],sort=False)
wins.head(20)
```

```python
ids = pd.concat([wins["champ1"],wins["champ2"],wins["champ3"],wins["champ4"],wins["champ5"]])
sort_champs = pd.DataFrame(ids,columns=["id"])
sort_champs = sort_champs.groupby(["id"])["id"].count().reset_index(name = "wins").set_index('id')
sort_champs
```

```python
champs_wins = champs.join(sort_champs)
champs_wins
```

```python
main_role = []
for id, items in champs_wins['tags'].items():
    main_role.append(items[0])
champs_wins["main_role"] = main_role
```

```python
winners = {
    "ids":[],
    "names":[],
    "roles":[],
    "titles":[],
    "wins":[]
}
for role in champs_wins["main_role"].unique():
    maxi = champs_wins[champs_wins["main_role"] == role]["wins"].max()
    values = champs_wins[champs_wins["wins"] == maxi].values[0]
    winners["names"].append(values[0])
    winners["roles"].append(values[5])
    winners["titles"].append(values[3])
    winners["wins"].append(values[4])
df = pd.DataFrame()
df["names"] = winners["names"]
df["roles"] = winners["roles"]
df["titles"] = winners["titles"]
df["wins"] = winners["wins"]
df.sort_values("wins",ascending=False).reset_index(drop=True)
```
